<h1>4: Prepare the data to better expose the underlying data patterns to machine learning algorithms

<h4>First, let's have a look at the missing values in terms of total bedroooms

```python
no_bedrooms = train[train["total_bedrooms"].isnull()]
housing_no_null = train.dropna(subset=["total_bedrooms"],axis=0)
ax = housing_no_null.plot(kind="scatter", x="longitude", y="latitude", alpha=0.1)
no_bedrooms.plot(ax=ax, color="r", kind="scatter", x="longitude", y="latitude", alpha=0.1)
plt.show()
```

```python
no_bedrooms.plot(kind="scatter", 
                 x="longitude", 
                 y="latitude", 
                 s=no_bedrooms["population"]/100, 
                 label="population",figsize=(10,7), 
                 alpha=0.4, c="median_house_value", 
                 cmap=plt.get_cmap("jet"), colorbar=True)
plt.show()
```

<h3>Ideally we'd have removed the rows with null values because they are fairly spread out and it would only decrese our dataset with 1%, but since the predictions on the test set has an expected amount of rows we'll go ahead and add the median value instead of all nulls

```python
housing_train = train.drop("median_house_value", axis=1)
housing_labels = train["median_house_value"]

housing_num = housing_train.drop("ocean_proximity", axis=1)
```

<h4>And let's build our pipeline for transforming and standardizing our dataset to perform better with our model:

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('std_scaler', StandardScaler()),
])

num_attributes = list(housing_num)
cat_attributes = ['ocean_proximity']

ohe = OneHotEncoder()
ohe.fit_transform(housing_train.dropna(axis=0))
ocean_prox_cat_1hot = ohe.categories_[8:]

full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_attributes),
    ('cat', OneHotEncoder(categories=ocean_prox_cat_1hot), cat_attributes)
])
```

```python
housing_prepared = full_pipeline.fit_transform(train)
```

```python
housing_prepared.shape
```

```python
housing_prepared[0]
```

<h4>The dataset looks okay, let's move on...

<h1>5: Explore many different models and short-list the best ones


<h4>Let's start out with a Linear regression model:

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(housing_prepared, housing_labels)
```

```python
from sklearn.metrics import mean_squared_error
housing_predictions = lin_reg.predict(housing_prepared)
lin_mse = mean_squared_error(housing_labels, housing_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
```

<h4>Now let's now train a decission tree regressor

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor()
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(tree_reg, 
                         housing_prepared, 
                         housing_labels, 
                         scoring="neg_mean_squared_error", 
                         cv=10)
tree_rmse_scores = np.sqrt(-scores)
display_scores(tree_rmse_scores)
```

<h4>let's have a look at the cross validation score:

```python
lin_scores = cross_val_score(lin_reg, 
                         housing_prepared, 
                         housing_labels, 
                         scoring="neg_mean_squared_error", 
                         cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)
display_scores(lin_rmse_scores)
```

<h4>This result from the tree model is not better than the linear model. Let's try another model

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor()
forest_reg.fit(housing_prepared, housing_labels)
```

```python
forest_predict = forest_reg.predict(housing_prepared)
forest_mse = mean_squared_error(housing_labels, forest_predict)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```

```python
scores = cross_val_score(forest_reg, 
                         housing_prepared, 
                         housing_labels, 
                         scoring="neg_mean_squared_error", 
                         cv=10)
forest_rmse_scores = np.sqrt(-scores)
display_scores(forest_rmse_scores)
```

<h4>SVR is another model that we should check out

```python
from sklearn.svm import SVR

svr = SVR(kernel='rbf', gamma='scale', C=300000)
svr.fit(housing_prepared, housing_labels)
svr_pred = svr.predict(housing_prepared)
svr_mse = mean_squared_error(housing_labels, svr_pred)
svr_rmse = np.sqrt(svr_mse)
svr_rmse
```

```python
svr = SVR(kernel='linear', gamma='scale', C=300000)
svr.fit(housing_prepared, housing_labels)
svr_pred = svr.predict(housing_prepared)
svr_mse = mean_squared_error(housing_labels, svr_pred)
svr_rmse = np.sqrt(svr_mse)
svr_rmse
```

<h4>Not too bad either with the rbf kernel on a Suport Vector regressor! A final idea is to try a Gradient Boosting regressor and combine that with the Random Forrest regressor in a Voting regressor:

```python
from sklearn.ensemble import GradientBoostingRegressor

gbr_reg = GradientBoostingRegressor()
gbr_reg.fit(housing_prepared, housing_labels)
gbr_pred = gbr_reg.predict(housing_prepared)
gbr_mse = mean_squared_error(housing_labels, gbr_pred)
gbr_rmse = np.sqrt(gbr_mse)
gbr_rmse
```

```python
scores = cross_val_score(gbr_reg, 
                         housing_prepared, 
                         housing_labels, 
                         scoring="neg_mean_squared_error", 
                         cv=10)
gbr_reg_rmse_scores = np.sqrt(-scores)
display_scores(gbr_reg_rmse_scores)
```

```python
from sklearn.ensemble import VotingRegressor

voting_reg = VotingRegressor(estimators=[('gb', gbr_reg), ('rf', forest_reg)])
voting_reg.fit(housing_prepared, housing_labels)
voting_pred = voting_reg.predict(housing_prepared)
voting_mse = mean_squared_error(housing_labels, voting_pred)
voting_rmse = np.sqrt(voting_mse)
voting_rmse
```

```python
scores = cross_val_score(voting_reg, 
                         housing_prepared, 
                         housing_labels, 
                         scoring="neg_mean_squared_error", 
                         cv=10)
voting_reg_rmse_scores = np.sqrt(-scores)
display_scores(voting_reg_rmse_scores)
```

<h4>It looks like the Random Forest Regressor and the Gradient Boosting Regressor performs fairly alike and we're having good results combining them in the Voting regressor. Let's work with this setup and see if we can finetune it. The SVR model is also performing well, but it takes long to train and has already been atempted tuned a bit. 

<h1>6: Fine-tune your models and combine them into a great solution


```python
from sklearn.model_selection import GridSearchCV

param_grid = [
    {'n_estimators':[8,10,50,100,300], 'max_features':[4,6,8]},
]

forest_reg = RandomForestRegressor()

grid_search = GridSearchCV(forest_reg, param_grid, cv=5, scoring='neg_mean_squared_error', return_train_score=True)

grid_search.fit(housing_prepared, housing_labels)
```

```python
grid_search.best_params_
```

```python
forest_reg = RandomForestRegressor(max_features=8, n_estimators=300)
forest_reg.fit(housing_prepared, housing_labels)
forest_predict = forest_reg.predict(housing_prepared)
forest_mse = mean_squared_error(housing_labels, forest_predict)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```

```python
scores = cross_val_score(forest_reg, 
                         housing_prepared, 
                         housing_labels, 
                         scoring="neg_mean_squared_error", 
                         cv=10)
forest_rmse_scores = np.sqrt(-scores)
display_scores(forest_rmse_scores)
```

<h4>Pretty good results even though at first it still looks like the model still is overfitting the data. Let's try and optimize the Gradient Boosting Regressor as well:

```python
param_grid = [
    {'n_estimators':[20,80,500,1500], 'random_state':[None, 1]},
]

gb_reg = GradientBoostingRegressor()

grid_search = GridSearchCV(gb_reg, param_grid, cv=5, scoring='neg_mean_squared_error', return_train_score=True)

grid_search.fit(housing_prepared, housing_labels)
```

```python
grid_search.best_params_
```

```python
gb_reg = GradientBoostingRegressor(n_estimators=1500, random_state=1)

gb_reg.fit(housing_prepared, housing_labels)
gb_pred = gb_reg.predict(housing_prepared)
gb_mse = mean_squared_error(housing_labels, gb_pred)
gb_rmse = np.sqrt(gb_mse)
gb_rmse
```

```python
scores = cross_val_score(gb_reg, 
                         housing_prepared, 
                         housing_labels, 
                         scoring="neg_mean_squared_error", 
                         cv=10)
gb_reg_rmse_scores = np.sqrt(-scores)
display_scores(gb_reg_rmse_scores)
```

```python
voting_reg = VotingRegressor(estimators=[('gb', gb_reg), ('rf', forest_reg)])
voting_reg.fit(housing_prepared, housing_labels)
voting_pred = voting_reg.predict(housing_prepared)
voting_mse = mean_squared_error(housing_labels, voting_pred)
voting_rmse = np.sqrt(voting_mse)
voting_rmse
```

```python
scores = cross_val_score(voting_reg, 
                         housing_prepared, 
                         housing_labels, 
                         scoring="neg_mean_squared_error", 
                         cv=10)
voting_reg_rmse_scores = np.sqrt(-scores)
display_scores(voting_reg_rmse_scores)
```

<h1>7: Present your solution</h1>
<p>We've teste and fine-tuned the models and tried different combinations. The voting regressor appears to be the most accurate model, though the RMSE score is a bit off compared to the cross validation scores. However, the fact that the voting regressor is a combination of the random forest regressor and the gradient boosting regressor seems to work in that it evens out the errors on both of the two others. 

```python
test_prepared = full_pipeline.fit_transform(test)
```

```python
result_voting_reg = voting_reg.predict(test_prepared)
result = pd.DataFrame(result_voting_reg)
result.columns = ['median_house_value']
result.to_csv('./data/submission_voting2.csv', index_label="Id")
len(result)
```
